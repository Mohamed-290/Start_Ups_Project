```python
x =data.iloc[:, 0:4]
y =data.iloc[:, 4:5]
```

```python
x = pd.get_dummies(x , drop_first=True)
```

```python
x
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.head ()
```

```python
data.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
x = scaler_x.fit_transform(x)
y = scaler_y.fit_transform(y)

from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 )
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train , y_train)
print("Train Score" ,model.score(x_test , y_test))
print("Test Score" ,model.score(x_train , y_train))
```

```python
from sklearn.svm import SVR
model = SVR()
model.fit(x_train , y_train)
print("Train Score :" ,model.score(x_test , y_test))
print("Test Score :" ,model.score(x_train , y_train))
```

## use kfold with SVR


exist error there careful

```python
from sklearn.model_selection import cross_val_score
model2 = SVR(kernel='linear', C=2)
scores = cross_val_score(model2 , x , y , cv=5)
print(scores)
print(scores.mean())

```

```python
from sklearn.neighbors import KNeighborsRegressor
model3 = KNeighborsRegressor(n_neighbors=3)
model3.fit(x_train, y_train)
print("Train Score :", model3.score(x_train, y_train))
print("Test Score :", model3.score(x_test, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
model4 = DecisionTreeRegressor()
model4.fit(x_train, y_train)
print("Train Score :", model4.score(x_train, y_train))
print("Test Score :", model4.score(x_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
model5 = RandomForestRegressor(n_estimators=150)
model5.fit(x_train, y_train)
print("Train Score :", model5.score(x_train, y_train))
print("Test Score :", model5.score(x_test, y_test))
```

```python
from xgboost import XGBRegressor
model6 = XGBRegressor()
model6.fit(x_train, y_train)
print("Train Score :", model6.score(x_train, y_train))
print("Test Score :", model6.score(x_test, y_test))
```
